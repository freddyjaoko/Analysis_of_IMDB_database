1. Listing all the directors who directed a 'Comedy' movie in a leap year. It should return director name, the movie name, and the year. 

```python
#finding the index of the row with the value "comedy" in the "Genre" column

genres = pd.read_sql_query("SELECT * FROM Genre LIMIT 20", conn)
genres

#comedy genre id=10
```

```python
#finding movies made in a leap year

leapyear = pd.read_sql_query("SELECT title, MID, Year FROM Movie WHERE (Year % 4 = 0 AND Year % 100 != 0) OR (Year % 400 = 0)", conn)
leapyear
```

```python
#finding movie id of movies in the comedy genre

comedy =pd.read_sql_query("SELECT MID FROM M_Genre WHERE GID = 10", conn)
comedy
```

```python
#merging the comedy and leapyear dataframes on "MID" to get a list of all the comedy movies released in a leap year

df_comedy = pd.merge(leapyear,comedy,on="MID", how= 'inner')
df_comedy
```

```python
#loading the directors id in a dataframe

directors = pd.read_sql_query("SELECT MID, PID FROM M_director", conn)
directors
```

```python
#merging df_comedy and directors dataframes on "PID"

df_directors = pd.merge(df_comedy,directors,on="MID", how= 'inner')
df_directors
```

```python
#merging the movie and director names data frames

director_names = pd.read_sql_query("SELECT PID, Name FROM Person", conn)
director_names
```

```python
#merging df_direcors and directors_names to get a dataframe of all the comedy movies made in a leapyear and their directors

df_directorsnames = pd.merge(df_directors, director_names, on='PID')

df_directorsnames
```

2. Listing the names of all the actors who played in the movie 'Anand' (1971)

3. List all the actors who acted in a film before 1970 and in a film after 1990. 

```python
#Listing all the movies made < 1970 and >1990
movies = pd.read_sql_query("SELECT MID FROM Movie WHERE Year < 1970 OR YEAR > 1990", conn)
movies
```

```python
#Listing all the cast members (actors) who were part of those movies < 1970 and >1990

cast_id = pd.read_sql_query("SELECT MID, PID FROM M_Cast", conn)

movies_cast = pd.merge(movies, cast_id, on='MID', how='inner')

cast_names = pd.read_sql_query("SELECT Name, PID FROM Person", conn)

m_cast_names = pd.merge(movies_cast, cast_names, on ='PID' )
m_cast_names
```

4. List all directors who directed 10 movies or more. Return the directors' names and the number of movies each of them directed.

```python

m_director = pd.read_sql_query("SELECT PID, MID FROM M_Director", conn)

person = pd.read_sql_query("SELECT Name, PID FROM Person", conn)

    
```

```python
#Counting the number of times a PID appears. This lets us know the number of movies a director has directed.
pid_10 = m_director['PID'].value_counts()

#Filtering to get the directors who have directed more than 10 movies
pid_10 = pid_10[pid_10 >= 10]

#Merging the person and the pid_10 dataframe to get the name of the directors and the number of directed movies.
m_directors = pd.merge(person, pid_10, on='PID', how='inner')
m_directors
```
